```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import sklearn
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
data.head()
```

```python
data.species.unique()
```

```python
sns.pairplot(data, hue = "species")
```

```python
y  = data["species"] # STRING BASED VALUES NEED TO BE CONVERTED TO NUMERIC
```

```python
categories = {'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3}
y_categorical = y.replace(categories)
```

```python
y_categorical
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(data[["sepal_h","sepal_w"]], y_categorical)
```

```python
model.predict([[5,2]])
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[["sepal_h","sepal_w"]], y_categorical, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred)*100,"%")
```
