```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import random
```

```python
x = np.array(range(-10,30))
y = np.array(range(-10,30))
```

```python
plt.scatter(x,y, color="r")
plt.plot(x,y, color="k")
plt.grid()
```

```python
df = pd.DataFrame(dict(x=x, y=y))
```

```python
df.corr()
```

```python
x = np.array(range(-10,30))
y = np.linspace(5,30,40)

plt.scatter(x,y, color="r")
plt.plot(x,y, color="k")
plt.yticks(y)
plt.grid()
```

```python
x = np.array(range(-10,30))
y = np.array(range(-10,30))
new_y = []
for index, value in enumerate(np.linspace(1,2, 40)):
    new_y.append(y[index]*value)
    
# new_y


plt.scatter(x,new_y, color="r")
plt.plot(x,new_y, color="k")
plt.grid()
```

```python
y
```

```python
df = pd.DataFrame(dict(x=x, y=new_y))
df.corr()
```

```python
x = np.array(range(-10,30))
y = [random.randint(10, 30) for _ in range(40)]

plt.scatter(x,y, color="r")
plt.plot(x,y, color="k")
plt.grid()
```

```python
df = pd.DataFrame(dict(x=x, y=y))
df.corr()
```

```python
x = np.array(range(-10,30))
y = [i*0.2+2 for i in x]

plt.scatter(x,y, color="r")
plt.plot(x,y, color="k")
plt.grid()
```
