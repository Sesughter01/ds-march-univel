### MAKE PYMYSQL INSERT OPERATION INTO A FUNCTION FOR EASE OF PLUGGABILITY INTO OUR CSV CODE.

```python
import pymysql.cursors

def write_to_db( customer, amount, collectionDate, tenor ):
    
    # Connect to the database
    connection = pymysql.connect(host='localhost',
                                 user='root',
                                 password='',
                                 database='loan',
                                 cursorclass=pymysql.cursors.DictCursor)
    with connection:
        with connection.cursor() as cursor:
            # Create a new record
            sql = "INSERT INTO `disbursements` (customer, amount, collectionDate, tenor) VALUES ('{0}', '{1}', '{2}', '{3}')"
            print(sql.format(customer, amount, collectionDate, tenor))
            cursor.execute(sql.format(customer, amount, collectionDate, tenor))

        connection.commit()
```
