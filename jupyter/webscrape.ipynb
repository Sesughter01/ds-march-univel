```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://www.jumia.com.ng/televisions/") # FETCH RAW DATA FROM WEBSITE
raw_html = response.text  # RETRIEVE RAW DATA IN ITS HTML FORMAT
raw_html
```

```python
parsed_data = BeautifulSoup(raw_html) # PARSE RAW DATA INTO A TRAVERSEABLE BEAUTIFULSOUP OBJECT
```

```python
parsed_data.find_all("div", {"class":"name"}) # find all divs with the class of name.
```

```python
found_data = parsed_data.find_all("div", {"class":"name"})
# Loop through all the divs with the class of name and then print their text

for data in found_data:
    print(data.text)
```

# Get names and prices too

```python
found_data = parsed_data.find_all("a", {"class":"core"})
found_data
```

```python
# Loop through all the divs with the class of name and then print their text

for data in found_data:
    name = data.find("div", {"class":"name"})
    price = data.find("div", {"class":"prc"})
    print(name.text)
    print(price.text, "\n")
```

### WEB SCRAPE ASSIGNMENT.

```python
response = requests.get("https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory") # FETCH RAW DATA FROM WEBSITE
raw_html = response.text  # RETRIEVE RAW DATA IN ITS HTML FORMAT
# raw_html
```

```python
parsed_data = BeautifulSoup(raw_html) 
```

```python
main_table = parsed_data.find("div", {"class":"scroll-container"})
```

```python
all_rows = main_table.find_all("tr")
# all_rows
```

```python
for row in all_rows:
    data = row.find_all("td")
    data_country_name = row.find("th")
    
    if len(data)<1: continue;
    
    print(data_country_name.text.strip().ljust(30), data[1].text.strip().ljust(10), data[2].text.strip().ljust(10), data[3].text.strip().ljust(10), "\n\n")
```
