```python
#Automate process getting all different unique Product lines in different sheets
```

```python
female_df = pd.read_excel("DataSets/m_f_sales.xlsx", sheet_name = "female_sales")
male_df = pd.read_excel("DataSets/m_f_sales.xlsx", sheet_name = "male_sales")
```

```python
# Get Unique Product lines
```

```python
female_df["Product line"].unique()
```

```python
male_df["Product line"].unique()
```

```python
unique_products_lines = female_df["Product line"].unique()

for prod_lin in unique_products_lines:
    
    print(prod_lin)
```

```python
unique_products_lines = female_df["Product line"].unique()

with pd.ExcelWriter('prod_lines.xlsx', mode='w') as writer: 
    
    for prod_lin_from_forloop in unique_products_lines:
        
        prod_line_mask = female_df["Product line"] == prod_lin_from_forloop

        prod_line_df = female_df[prod_line_mask]
        prod_line_df.to_excel(writer, sheet_name=prod_lin_from_forloop)
```
