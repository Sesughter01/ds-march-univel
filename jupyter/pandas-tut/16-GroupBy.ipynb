# Tutorial 12: Groupby

## What is GroupBy?

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(
    {"key1":list("aabbab"),
     "key2":["one","two","three"]*2,
     "data1":np.random.randn(6),
     "data2":np.random.randn(6)})
df
```

```python
group=df.groupby(by = "key1")
```

```python
group
```

```python
group.mean()
```

```python
ave=df.groupby([df["key1"],
                         df["key2"]]).mean()
ave
```

```python
ave.unstack()
```

```python
df.groupby("key1").mean()
```

```python
df.groupby(["key1","key2"]).mean()
```

## Iterating over Groups

```python
for name, group in df.groupby("key1"):
    print(name)
    print(group)
```

```python
for (x1,x2),group in df.groupby(["key1",
                                 "key2"]):
    print(x1,x2)
    print(group)
```

```python
piece=dict(list(df.groupby("key1")))
```

```python
piece["a"]
```

## Selecting a Column or Subset of Columns

```python
df.groupby(['key1', 
            'key2'])[['data1']].mean()
```

## Grouping with Dicts and Series

```python
fruit=pd.DataFrame(np.random.randn(4,4),
                   columns=list("abcd"),
                   index=["apple","cherry",
                          "banana","kiwi"])
fruit
```

```python
label={"a": "green","b":"yellow",
       "c":"green","d":"yellow",
       "e":"purple"}
```

```python
group=fruit.groupby(label,axis=1)
```

```python
group.sum()
```

```python
s=pd.Series(label)
s
```

```python
fruit.groupby(s,axis=1).count()
```

## Grouping with Functions

```python
fruit.groupby(len).sum()
```

## Grouping by Index Levels

```python
data=pd.DataFrame(np.random.randn(4,5),
                  columns=[list("AAABB"),
                           [1,2,3,1,2]])
```

```python
data.columns.names=["letter","number"]
data
```

```python
data.groupby(level="letter",axis=1).sum()
```
