```python
# Multiple plots at once (Playing with the canvas)
"""
Three key components in matplotlib's operations

1- Artists -> Every visual point, lines, markers, etc
2- Canvas -> Boundary in which artists can draw or express themselves
3- Figure -> Is the entirety of area that canvases can sit on.

"""
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x,y, color = 'r', marker="v", label="Ghana") # Scatter plot with title, labels, and legend
plt.plot(x,y2, color = 'b', marker="v", label="Nigeria") # Scatter plot with title, labels, and legend
plt.title("RANDOM PLOT")
plt.xlabel("Sales")
plt.ylabel("Quantity")
plt.legend()
plt.show()
```

```python
# To create multiple plots or canvases in a figure, the subplots feature of the figure has to be invoked and configureed

fig, ax = plt.subplots(1, 3 ,figsize=(15,3)) # plt.subplots(ROW, COL), figsize=(width,height) 
# fig represents the main figure
# ax represents the number of canvases
fig.suptitle("MAIN CHART")
plt.plot()
ax
```

```python
fig, ax = plt.subplots(1, 3 ,figsize=(15,3)) # plt.subplots(ROW, COL)
# fig represents the main figure
# ax represents the number of canvases

# fig.suptitle("MAIN CHART")
ax[0].plot(x,y, color = 'r', marker="v", label="Ghana") # Scatter plot with title, labels, and legend
ax[1].bar(x,y, color = 'r', label="Ghana")
ax[2].pie([2,2,2], labels = ["Charles", "Ridwan", "Ayo"]) # Pie chart with labels

ax[0].set_title("Line Chart")
ax[1].set_title("Bar Graph")
ax[2].set_title("Pie Chart")

plt.plot()
```

```python
fig, ax = plt.subplots(2, 3 ,figsize=(15,8)) # 2 ROWS THIS TIME
# fig represents the main figure
# ax represents the number of canvases

# fig.suptitle("MAIN CHART")
ax[0,0].plot(x,y, color = 'r', marker="v", label="Ghana") # Scatter plot with title, labels, and legend
ax[0,1].bar(x,y, color = 'r', label="Ghana")
ax[0,2].pie([2,2,2], labels = ["Charles", "Ridwan", "Ayo"]) # Pie chart with labels

ax[1,2].plot(x,y, color = 'r', marker="v", label="Ghana") # Scatter plot with title, labels, and legend
ax[1,1].bar(x,y, color = 'r', label="Ghana")
ax[1,0].pie([2,2,2], labels = ["Charles", "Ridwan", "Ayo"]) # Pie chart with labels

# ax[0].set_title("Line Chart")
# ax[1].set_title("Bar Graph")
# ax[2].set_title("Pie Chart")

# plt.plot()
ax
```

```python
'''
==============
3D scatterplot
==============

Demonstration of a basic scatterplot in 3D.
'''

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
